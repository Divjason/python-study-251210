```python
import requests
import openpyxl
import re

client_id = ""
client_secret = ""
start = 1
num = 0

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions["B"].width = 100
excel_sheet.column_dimensions["C"].width = 100
excel_sheet.append(["No", "ì œëª©", "ë§í¬"])

for index in range(10) :
    start_number = start + (index * 10)
    naver_open_api = "https://openapi.naver.com/v1/search/shop.json?query=ë¡±íŒ¨ë”©&display=100&start=" + str(start_number)
    header_params = {"X-Naver-Client-Id": client_id, "X-Naver-Client-Secret": client_secret}
    
    res = requests.get(naver_open_api, headers=header_params)

    if res.status_code == 200 :
        data = res.json()
        for item in data["items"] :
            num += 1
            clean_title = re.sub(r"<.*?>", "", item["title"])
            excel_sheet.append([num, clean_title, item["link"]])
    else :
        print("Error Code : ", res.status_code)
        
    excel_file.save("shopping_with_ë¡±íŒ¨ë”©.xlsx")
    excel_file.close()

print("ğŸ shopping_with_ë¡±íŒ¨ë”©.xlsx ì €ì¥ì™„ë£Œ")
```
