```python
# crawl : 기어가다
# crawling : 기어가는 것 (행위)
# 이미 다른 사람이 만들어놓은 웹 사이트의 정보를 몰래가서 수집!!
# 네이버 사이트 데이터를 크롤링 하자!!
# crawler : 크롤링을 하는 주체 : requests 라이브러리 모듈
# pseudo code : 코드를 설계 할 때, 
# 셀 = cell

import requests
from bs4 import BeautifulSoup # Class

query = input("네이버뉴스 검색키워드 : ")
url = f"https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query={query}"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

titles = soup.select("a.fender-ui_228e3bd1.moM44hE6Je7O8nL1iBI9")

for title in titles :
    print(title.text.strip())

# 3개의 주체!!!

# 사용자 user
# get 요청 (request)
# 브라우저 client
# response
# 서버 server
# [ ] => 리스트 = 배열 (array)
# 외부에서 값을 찾아오는 크롤링 행위 => 100% 모두 리스트 자료형태로 값을 찾아옴!!!
```
