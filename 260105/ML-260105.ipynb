```python
from sklearn.datasets import load_digits
digits = load_digits() # 컴퓨터에게 학습시킬 문제, 정답
digits
```

```python
digits["data"] # 실제 각 이미지를 숫자화한 데이터
```

```python
from matplotlib import pyplot as plt

%matplotlib inline
fig = plt.figure(figsize=(6, 6))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64) :
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation="nearest")
    ax.text(0, 7, str(digits.target[i]))

plt.show()
```

1. 지도학습 : 답 & 문제 : 학습 -> 문제
2. 비지도학습 : 문제 : 근거자료 -> 학습 // 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target)

# KNN
# Logistic Regression
# Decision Tree
# Random Forest
# SVM
# XGboost
# Light GBM
# GNB
# ....

# 단순히 모듈 => 정확도
```

```python
from sklearn.naive_bayes import GaussianNB

OneOfNaiveBayesianModels = GaussianNB()
```

```python
OneOfNaiveBayesianModels.fit(X_train, y_train)
```

```python
predicted = OneOfNaiveBayesianModels.predict(X_test)
```

```python
fig = plt.figure(figsize=(6, 6))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64) :
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(X_test.reshape(-1, 8, 8)[i], cmap=plt.cm.binary, interpolation="nearest")

    if predicted[i] == y_test[i] :
        ax.text(0, 7, str(predicted[i]), color="green")
    else :
        ax.text(0, 7, str(predicted[i]), color="red")

plt.show()
```

```python
# M/L VS D/L
# 학습데이터 > 모델(훈련) > 답
# 한계 
# 원숭이 엉덩이는 빨개 -> 빨가면 사과 -> 사과는 맛있어 -> 맛있으면 바나나
```
